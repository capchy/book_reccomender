```python
ratings = pd.read_csv("BX-Book-Ratings.csv",delimiter=';',on_bad_lines='skip',encoding='latin-1')
```

```python
ratings=ratings[ratings['user_id'].isin(y)]
```

```python
ratings_with_books=ratings.merge(books, on="ISBN")
```

```python
ratings_with_books.head(5)
```

```python
num_ratings= ratings_with_books.groupby('title')['rating'].count().reset_index()
```

```python
num_ratings
```

```python
num_ratings.rename(columns={'rating':'num_of_rating'},inplace =True)
```

```python
num_ratings.head(45)
```

```python
final_ratings=ratings_with_books.merge(num_ratings, on = 'title')
```

```python
final_ratings.head(2)
```

```python
final_ratings=final_ratings[final_ratings['num_of_rating']>35]
```

```python
final_ratings.head(2)
```

```python
final_ratings.shape
```

```python
final_ratings.drop_duplicates(['user_id','title'],inplace=True)
```

```python
book_pivot=final_ratings.pivot_table(columns='user_id',index='title',values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0, inplace =True)
```

```python
def recommend_book(book_name):
    book_id = np.where(book_pivot.index == book_name)[0][0]
    distance, suggestion = model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1), n_neighbors=6 )
    
    for i in range(len(suggestion)):
            books = book_pivot.index[suggestion[i]]
            for j in books:
                if j == book_name:
                    print(f"You searched '{book_name}'\n")
                    print("The suggestion books are: \n")
                else:
                    print(j)
```

```python
book_name = "Harry Potter and the Chamber of Secrets (Book 2)"
recommend_book(book_name)
```
